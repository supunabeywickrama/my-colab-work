```python
import shutil, yaml
from tqdm import tqdm

FINAL_CLASSES = [
    "elephant",
    "monkey",
    "cow",
    "buffalo",
    "peacock",
    "pig",
    "porcupine"
]

CLASS_IDS = {c:i for i,c in enumerate(FINAL_CLASSES)}

LABEL_MAP = {
    "elephant": "elephant",
    "monkey": "monkey",
    "cow": "cow",
    "cattle": "cow",
    "buffalo": "buffalo",
    "peacock": "peacock",
    "pig": "pig",
    "boar": "pig",
    "wild_boar": "pig",
    "porcupine": "porcupine"
}

MERGED = "/content/merged_dataset"
for d in ["images/train","labels/train"]:
    os.makedirs(os.path.join(MERGED,d), exist_ok=True)

def merge_dataset(base):
    yaml_path = os.path.join(base,"data.yaml")
    if not os.path.exists(yaml_path):
        return

    with open(yaml_path) as f:
        names = yaml.safe_load(f)["names"]

    for split in ["train","valid","val"]:
        img_dir = os.path.join(base,split,"images")
        lbl_dir = os.path.join(base,split,"labels")
        if not os.path.exists(img_dir):
            continue

        for img in os.listdir(img_dir):
            shutil.copy(
                os.path.join(img_dir,img),
                os.path.join(MERGED,"images/train",img)
            )

            lbl = os.path.splitext(img)[0]+".txt"
            src_lbl = os.path.join(lbl_dir,lbl)
            if not os.path.exists(src_lbl):
                continue

            with open(src_lbl) as f:
                lines = f.readlines()

            new_lines = []
            for l in lines:
                cid,*box = l.strip().split()
                cname = LABEL_MAP.get(names[int(cid)].lower())
                if cname in CLASS_IDS:
                    new_lines.append(f"{CLASS_IDS[cname]} "+" ".join(box))

            if new_lines:
                with open(os.path.join(MERGED,"labels/train",lbl),"w") as f:
                    f.write("\n".join(new_lines))

for p in tqdm(paths):
    merge_dataset(p)
```

```python
import os, random, shutil
from collections import defaultdict

SRC_IMG = "/content/merged_dataset/images/train"
SRC_LBL = "/content/merged_dataset/labels/train"

DST = "/content/filtered_dataset"
DST_IMG = f"{DST}/images/train"
DST_LBL = f"{DST}/labels/train"

os.makedirs(DST_IMG, exist_ok=True)
os.makedirs(DST_LBL, exist_ok=True)

FINAL_CLASSES = [
    "elephant",
    "monkey",
    "cow",
    "buffalo",
    "peacock",
    "pig",
    "porcupine"
]

CLASS_LIMITS = {
    "elephant": 1000,
    "buffalo": 1000,
    "cow": 1000
}

# map class id -> class name
id2name = {i:n for i,n in enumerate(FINAL_CLASSES)}

# collect images per class
class_images = defaultdict(list)

for lbl in os.listdir(SRC_LBL):
    lbl_path = os.path.join(SRC_LBL, lbl)
    with open(lbl_path) as f:
        lines = f.readlines()

    used_classes = set()
    for l in lines:
        cid = int(l.split()[0])
        used_classes.add(id2name[cid])

    img = lbl.replace(".txt", ".jpg")
    if os.path.exists(os.path.join(SRC_IMG, img)):
        for c in used_classes:
            class_images[c].append(img)

# decide which images to keep
keep_images = set()

for cls, imgs in class_images.items():
    if cls in CLASS_LIMITS:
        keep_images.update(random.sample(imgs, CLASS_LIMITS[cls]))
    else:
        keep_images.update(imgs)

# copy filtered data
for img in keep_images:
    shutil.copy(os.path.join(SRC_IMG, img), os.path.join(DST_IMG, img))
    lbl = img.replace(".jpg", ".txt")
    shutil.copy(os.path.join(SRC_LBL, lbl), os.path.join(DST_LBL, lbl))

print("Filtered images:", len(keep_images))
```

```python
from collections import Counter
import os

counts = Counter()

# Use DST_LBL which points to the filtered dataset's labels
for lbl in os.listdir(DST_LBL):
    with open(os.path.join(DST_LBL,lbl)) as f:
        for l in f:
            counts[int(l.split()[0])] += 1

FINAL_CLASSES = [
    "elephant","monkey","cow",
    "buffalo","peacock","pig","porcupine"
]

for cid,c in counts.items():
    print(FINAL_CLASSES[cid],":",c)
```
