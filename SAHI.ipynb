```python
import os

base = "/content/dataset"

folders = [
    "images/train", "images/val",
    "labels/train", "labels/val"
]

for f in folders:
    os.makedirs(os.path.join(base, f), exist_ok=True)

print("Unified dataset folders created")
```

```python
import shutil
import glob
import os

datasets = [
    "/content/crowd-1",
    "/content/crowd-counting-dataset-2"
]

for ds in datasets:
    for split in ["train", "valid"]:
        img_src = f"{ds}/{split}/images"
        lbl_src = f"{ds}/{split}/labels"

        img_dst = f"/content/dataset/images/{'train' if split=='train' else 'val'}"
        lbl_dst = f"/content/dataset/labels/{'train' if split=='train' else 'val'}"

        if os.path.exists(img_src):
            for f in glob.glob(img_src + "/*"):
                shutil.copy(f, img_dst)

        if os.path.exists(lbl_src):
            for f in glob.glob(lbl_src + "/*"):
                shutil.copy(f, lbl_dst)

print("✅ Merge completed")

```
