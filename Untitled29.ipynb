```python
import cv2, numpy as np

# === constants & maps (same as before) ===
DIGIT = {'black':0,'brown':1,'red':2,'orange':3,'yellow':4,
         'green':5,'blue':6,'violet':7,'grey':8,'white':9}
MULT = {'black':1,'brown':10,'red':100,'orange':1e3,'yellow':1e4,
        'green':1e5,'blue':1e6,'violet':1e7,'grey':1e8,'white':1e9,'gold':0.1,'silver':0.01}
TOL  = {'brown':1.0,'red':2.0,'green':0.5,'blue':0.25,'violet':0.1,'grey':0.05,'gold':5.0,'silver':10.0}
COLOR_REF_LAB = {
    'black':(0,0,0),'brown':(37,14,14),'red':(53,80,67),'orange':(64,62,53),'yellow':(82,-5,88),
    'green':(46,-51,49),'blue':(32,79,-108),'violet':(29,58,-36),'grey':(53,0,0),'white':(100,0,0),
    'gold':(70,10,40),'silver':(78,0,0)
}

def _to_lab(bgr): return cv2.cvtColor(bgr, cv2.COLOR_BGR2LAB).astype(np.float32)
def _delta_e(a,b): return float(np.linalg.norm(np.array(a,np.float32)-np.array(b,np.float32)))

def _nearest_color_name(mean_lab):
    best, name = 1e9, None
    for k,v in COLOR_REF_LAB.items():
        d = _delta_e(mean_lab, v)
        if d < best: best, name = d, k
    return name, best

def _white_balance_grayworld(img):
    # simple gray-world white balance
    b,g,r = cv2.split(img.astype(np.float32))
    kb = (b.mean()+g.mean()+r.mean())/(3*b.mean()+1e-6)
    kg = (b.mean()+g.mean()+r.mean())/(3*g.mean()+1e-6)
    kr = (b.mean()+g.mean()+r.mean())/(3*r.mean()+1e-6)
    out = cv2.merge([b*kb, g*kg, r*kr]); out = np.clip(out,0,255).astype(np.uint8)
    return out

def _segment_bands_1d(lab_strip, min_frac=0.035):
    """Return list of (x1,x2) band spans from a LAB strip."""
    h, w = lab_strip.shape[:2]
    prof = np.median(lab_strip, axis=0)  # (w,3)
    diff = np.linalg.norm(np.diff(prof, axis=0), axis=1)
    thr = np.percentile(diff, 85)
    edges = np.where(diff > thr)[0]

    if edges.size < 3:
        return []

    # build candidate cuts
    cuts = [0]
    for e in edges:
        if cuts and (e - cuts[-1]) < 4:  # merge close edges
            continue
        cuts.append(int(e))
    cuts.append(w-1)

    bands = []
    minw = max(6, int(min_frac*w))
    for i in range(len(cuts)-1):
        x1, x2 = cuts[i], cuts[i+1]
        if (x2 - x1) >= minw:
            bands.append((x1, x2))

    # drop over-wide â€œbodyâ€ segments > 0.35 * width
    bands = [b for b in bands if (b[1]-b[0]) <= 0.35*w]
    return bands

def _dominant_color_lab(segment_lab):
    """k=2 k-means in LAB, return dominant cluster's mean + size."""
    pts = segment_lab.reshape(-1,3).astype(np.float32)
    if pts.shape[0] < 50:
        mean = np.median(pts, axis=0)
        return mean, 1.0
    # k-means
    criteria = (cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER, 15, 0.5)
    K=2
    compactness, labels, centers = cv2.kmeans(pts, K, None, criteria, 3, cv2.KMEANS_PP_CENTERS)
    counts = np.bincount(labels.flatten(), minlength=K)
    k = int(np.argmax(counts))
    frac = float(counts[k])/float(max(1,counts.sum()))
    return centers[k], frac

def read_resistor_value_improved(crop_bgr):
    # 1) pre-clean
    img = _white_balance_grayworld(crop_bgr)
    img = cv2.bilateralFilter(img, 7, 60, 60)

    # CLAHE on L channel
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    L,a,b = cv2.split(lab)
    L = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8)).apply(L)
    lab = cv2.merge([L,a,b]).astype(np.float32)

    # 2) central thin strip
    h,w = img.shape[:2]
    y1,y2 = int(0.45*h), int(0.55*h)  # thinner strip reduces body bleed
    strip_lab = lab[y1:y2, :, :]

    # 3) find band spans
    spans = _segment_bands_1d(strip_lab, min_frac=0.035)
    if len(spans) < 4:
        return None, "bands-not-found"

    # 4) classify each span via k-means dominant cluster
    colors, confid = [], []
    for (x1,x2) in spans:
        seg = strip_lab[:, x1:x2, :]
        mean_lab, frac = _dominant_color_lab(seg)
        name, dE = _nearest_color_name(mean_lab)
        # reject very weak matches
        if dE > 30:  # Î”E gate; tune 25â€“35
            name = None
        colors.append(name)
        # crude confidence: weight by cluster dominance and Î”E
        c = max(0.0, 1.0 - dE/30.0) * frac
        confid.append(round(float(c),2))

    # filter invalids
    valid = [(c,sp) for c,sp in zip(colors,spans) if c is not None]
    if len(valid) < 4:
        return None, "low-color-confidence"
    colors = [c for c,_ in valid]

    # 5) orientation: tolerance likely gold/silver at RIGHT end
    def is_tol(c): return c in TOL
    if colors and colors[0] in ('gold','silver') and not (colors[-1] in ('gold','silver')):
        colors = list(reversed(colors))

    # keep best consecutive 4â€“5 band slice that ends with tolerance
    slices = []
    for k in (5,4):
        if len(colors) >= k:
            for i in range(len(colors)-k+1):
                chunk = colors[i:i+k]
                if chunk[-1] in TOL: slices.append(chunk)
    if not slices:
        # fallback: take the central 4/5 bands
        mid = len(colors)//2
        cand = colors[max(0,mid-2):mid+3]  # try 5
        if len(cand)<4: cand = colors[:4]
        slices = [cand]

    bands = slices[0][:5] if len(slices[0])>=5 else slices[0][:4]

    # 6) compute value
    if len(bands)==5:
        d1,d2,d3,mul,t = bands
        if d1 in DIGIT and d2 in DIGIT and d3 in DIGIT and mul in MULT and t in TOL:
            val = (100*DIGIT[d1]+10*DIGIT[d2]+DIGIT[d3])*MULT[mul]
            return (val, t), None
    if len(bands)==4:
        d1,d2,mul,t = bands
        if d1 in DIGIT and d2 in DIGIT and mul in MULT and t in TOL:
            val = (10*DIGIT[d1]+DIGIT[d2])*MULT[mul]
            return (val, t), None

    return None, "parse-failed"
```

```python
# ============================
# Robust resistor reader v3  (black/gray safe)
# ============================
import os, cv2, numpy as np, matplotlib.pyplot as plt
from typing import List, Tuple, Optional
from google.colab import files

# ---- YOLO model assumed loaded as `model` ----
CONF_THRES  = 0.20
PAD_FRAC    = 0.18          # pad YOLO box so bands aren't truncated
UPSCALE_MIN = 320           # upsize tiny crops
DE_GATE     = 55            # Î”E tolerance (more permissive)
SAVE_DIR = "/content/output"; os.makedirs(SAVE_DIR, exist_ok=True)

# ---- color maps ----
COLOR_REF_LAB = {
    'black':(0,0,0),'brown':(37,14,14),'red':(53,80,67),'orange':(64,62,53),
    'yellow':(82,-5,88),'green':(46,-51,49),'blue':(32,79,-108),'violet':(29,58,-36),
    'gray':(53,0,0),'white':(100,0,0),'gold':(70,10,40),'silver':(78,0,0)
}
DIGIT = {'black':0,'brown':1,'red':2,'orange':3,'yellow':4,'green':5,'blue':6,'violet':7,'gray':8,'white':9}
MULT  = {'black':1,'brown':10,'red':100,'orange':1e3,'yellow':1e4,'green':1e5,'blue':1e6,'violet':1e7,'gray':1e8,'white':1e9,'gold':0.1,'silver':0.01}
TOL   = {'brown':1.0,'red':2.0,'green':0.5,'blue':0.25,'violet':0.1,'gray':0.05,'gold':5.0,'silver':10.0}
TEMP_COEFF = {'brown':100,'red':50,'orange':15,'yellow':25,'blue':10,'violet':5}

# ---- helpers ----
def fmt_ohms(v):
    if v is None: return "?"
    if v>=1e6: return f"{v/1e6:.2g} MÎ©"
    if v>=1e3: return f"{v/1e3:.2g} kÎ©"
    return f"{v:.0f} Î©"

def delta_e(a,b):
    a=np.array(a,np.float32); b=np.array(b,np.float32); return float(np.linalg.norm(a-b))

def nearest_color_name(lab):
    best=1e9; name=None
    for k,v in COLOR_REF_LAB.items():
        d=delta_e(lab,v)
        if d<best: best, name = d, k
    return name, best

def grayworld(bgr):
    b,g,r = cv2.split(bgr.astype(np.float32))
    m=(b.mean()+g.mean()+r.mean())/3.0
    out=cv2.merge([b*m/(b.mean()+1e-6), g*m/(g.mean()+1e-6), r*m/(r.mean()+1e-6)])
    return np.clip(out,0,255).astype(np.uint8)

def gamma_corr(img,g=1.0):
    lut=np.array([(i/255.)**g*255 for i in range(256)],np.float32).clip(0,255).astype(np.uint8)
    return cv2.LUT(img,lut)

def deskew_minarea(bgr):
    """Deskew using largest contour's minAreaRect (ignores thin leads)."""
    g=cv2.cvtColor(bgr,cv2.COLOR_BGR2GRAY)
    g=cv2.GaussianBlur(g,(5,5),0)
    thr=cv2.adaptiveThreshold(g,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,51,5)
    kernel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(7,7))
    mask=cv2.morphologyEx(thr,cv2.MORPH_CLOSE,kernel,iterations=2)
    cnts,_=cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    if not cnts:
        return bgr, None
    cnt=max(cnts,key=cv2.contourArea)
    rect=cv2.minAreaRect(cnt)
    angle=rect[2]
    if angle<-45: angle += 90
    (h,w)=bgr.shape[:2]
    M=cv2.getRotationMatrix2D((w//2,h//2), angle, 1.0)
    rot=cv2.warpAffine(bgr,M,(w,h),flags=cv2.INTER_LINEAR,borderMode=cv2.BORDER_REPLICATE)
    return rot, rect

def body_roi(rot, rect):
    """Crop tight box around body in rotated image, expand slightly."""
    if rect is None:
        return rot
    (cx,cy),(rw,rh),_ = rect
    if rh>rw: rw,rh = rh,rw  # want long side horizontal after rotation
    pad=0.18
    x1=int(max(0,cx - (rw*(1+pad))/2))
    x2=int(min(rot.shape[1], cx + (rw*(1+pad))/2))
    y1=int(max(0,cy - (rh*(1+pad))/2))
    y2=int(min(rot.shape[0], cy + (rh*(1+pad))/2))
    return rot[y1:y2, x1:x2]

def segment_bands(strip_bgr):
    """
    Segment with BOTH color-edge (Lab Î”) and intensity valleys (low L) â†’ works for black bands.
    """
    h,w = strip_bgr.shape[:2]
    lab_u8 = cv2.cvtColor(strip_bgr, cv2.COLOR_BGR2LAB)
    L,a,b = cv2.split(lab_u8)
    Lc = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8)).apply(L)

    # color edges
    lab = np.dstack([Lc,a,b]).astype(np.float32)
    prof = np.median(lab, axis=0)      # (w,3)
    diff = np.linalg.norm(np.diff(prof, axis=0), axis=1)
    diff = cv2.GaussianBlur(diff.reshape(1,-1),(1,9),0).ravel()

    # darkness valleys
    Lprof = np.median(Lc, axis=0).astype(np.float32)
    Lz = (Lprof - Lprof.mean())/ (Lprof.std()+1e-6)
    dark = (-Lz).clip(0)

    # combine - Slice dark to match diff length
    score = (diff/(diff.max()+1e-6))*0.6 + (dark[:len(diff)]/(dark.max()+1e-6))*0.4
    thr = max(0.15, float(np.percentile(score, 68)))
    peaks = np.where(score>thr)[0]
    if peaks.size<3:
        return []

    cuts=[0]; last=-999
    for p in peaks:
        if p-last<5: continue
        cuts.append(int(p)); last=p
    cuts.append(w-1)

    minw = max(6, int(0.026*w))
    maxw = int(0.50*w)
    spans=[]
    for i in range(len(cuts)-1):
        x1,x2 = cuts[i], cuts[i+1]
        ww = x2-x1
        if ww>=minw and ww<=maxw:
            spans.append((x1,x2))
    return spans

def classify_band(seg_bgr):
    """Return color name and Î”E; explicit black/gray handling."""
    lab_u8 = cv2.cvtColor(seg_bgr, cv2.COLOR_BGR2LAB)
    mean_lab = np.median(lab_u8.reshape(-1,3), axis=0).astype(np.float32)
    hsv = cv2.cvtColor(seg_bgr, cv2.COLOR_BGR2HSV)
    V = float(np.median(hsv[...,2])); S = float(np.median(hsv[...,1]))
    L = float(np.median(lab_u8[...,0]))

    # explicit black
    if V < 80 and L < 70:
        return 'black', 0.0
    # explicit gray (low sat, mid value)
    if S < 25 and 80 <= V <= 180 and 70 <= L <= 160:
        return 'gray', 0.0

    name, dE = nearest_color_name(mean_lab)
    # low-sat fallback to gold/silver if very bright/dull
    if S < 22 and name not in ('gold','silver','black','gray'):
        name = 'gold' if V>170 else 'silver'
        dE = 0.0
    return name, dE

def decode_bands(bands: List[str]):
    tight = set(['brown','red','green','blue','violet','gray','gold','silver'])
    if bands and (bands[0] in tight) and not (bands[-1] in tight):
        bands = list(reversed(bands))

    # try 6 â†’ 5 â†’ 4
    if len(bands)>=6:
        for i in range(len(bands)-5):
            d1,d2,d3,m,t,tc = bands[i:i+6]
            if all(x in DIGIT for x in (d1,d2,d3)) and m in MULT and t in TOL and tc in TEMP_COEFF:
                return (100*DIGIT[d1]+10*DIGIT[d2]+DIGIT[d3])*MULT[m], TOL[t], TEMP_COEFF[tc]
    if len(bands)>=5:
        for i in range(len(bands)-4):
            d1,d2,d3,m,t = bands[i:i+5]
            if all(x in DIGIT for x in (d1,d2,d3)) and m in MULT and t in TOL:
                return (100*DIGIT[d1]+10*DIGIT[d2]+DIGIT[d3])*MULT[m], TOL[t], None
    if len(bands)>=4:
        for i in range(len(bands)-3):
            d1,d2,m,t = bands[i:i+4]
            if all(x in DIGIT for x in (d1,d2)) and m in MULT and t in TOL:
                return (10*DIGIT[d1]+DIGIT[d2])*MULT[m], TOL[t], None
    # gentle fallback if tolerance uncertain but digits & multiplier look valid
    if len(bands)>=4:
        for i in range(len(bands)-3):
            d1,d2,m,_ = bands[i:i+4]
            if all(x in DIGIT for x in (d1,d2)) and m in MULT:
                return (10*DIGIT[d1]+DIGIT[d2])*MULT[m], None, None
    return None, None, None

def read_resistor_value(crop_bgr):
    if crop_bgr is None or crop_bgr.size==0:
        return None, "empty-crop", None

    H0,W0 = crop_bgr.shape[:2]
    img = gamma_corr(grayworld(crop_bgr), 1.05)
    if max(H0,W0) < UPSCALE_MIN:
        s = UPSCALE_MIN/max(H0,W0)
        img = cv2.resize(img, (int(W0*s), int(H0*s)), interpolation=cv2.INTER_CUBIC)

    rot, rect = deskew_minarea(img)
    roi = body_roi(rot, rect)
    if roi is None or roi.size==0: roi = rot
    h,w = roi.shape[:2]
    if h<12 or w<30:
        return None, "crop-too-small", None

    # middle strip; widen if needed
    for (yl,yh) in [(0.40,0.60),(0.35,0.65),(0.45,0.55)]:
        y1,y2 = int(yl*h), int(yh*h)
        strip = roi[y1:y2,:]
        spans = segment_bands(strip)
        if len(spans) >= 3:
            break

    if len(spans) < 3:
        return None, "bands-not-found", dict(rot=rot, roi=roi, strip=strip)

    # classify bands
    colors=[]; dEs=[]
    for (x1,x2) in spans:
        seg = strip[:, x1:x2]
        name, dE = classify_band(seg)
        # PERMISSIVE: accept black/gray even if Î”E big
        if name in ('black','gray'):
            colors.append(name); dEs.append(dE)
        elif dE > DE_GATE:
            colors.append(None); dEs.append(dE)
        else:
            colors.append(name); dEs.append(dE)

    # fallback: if mostly dark bands, fill Nones with 'black'
    valid = [c for c in colors if c is not None]
    if len(valid) < 4 and colors.count('black') >= 2:
        colors = [c if c is not None else 'black' for c in colors]
        valid = [c for c in colors if c is not None]

    if len(valid) < 4:
        return None, "low-color-confidence", dict(rot=rot, roi=roi, strip=strip, spans=spans, colors=colors)

    # try both orientations
    for cand in (valid, list(reversed(valid))):
        val, tol, tppm = decode_bands(cand)
        if val is not None:
            return (val, tol, tppm), None, dict(rot=rot, roi=roi, strip=strip, spans=spans, colors=cand)

    return None, "decode-failed", dict(rot=rot, roi=roi, strip=strip, spans=spans, colors=colors)

# ---------- RUN ON AN UPLOADED IMAGE ----------
uploaded = files.upload()
img_path = list(uploaded.keys())[0]

preds = model.predict(source=img_path, conf=CONF_THRES, verbose=False)
img = preds[0].orig_img.copy()
H,W = img.shape[:2]

for b in preds[0].boxes:
    x1,y1,x2,y2 = map(int, b.xyxy[0])
    # pad box
    pw = int((x2-x1)*PAD_FRAC); ph = int((y2-y1)*PAD_FRAC)
    x1 = max(0, x1-pw); y1 = max(0, y1-ph)
    x2 = min(W, x2+pw); y2 = min(H, y2+ph)
    if x2<=x1 or y2<=y1: continue

    crop = img[y1:y2, x1:x2]
    out, err, dbg = read_resistor_value(crop)

    if out is None:
        cv2.rectangle(img,(x1,y1),(x2,y2),(0,165,255),2)
        cv2.putText(img,f"Error: {err}",(x1,max(20,y1-10)),cv2.FONT_HERSHEY_SIMPLEX,0.7,(0,165,255),2)
    else:
        val, tol, tppm = out
        label = fmt_ohms(val) + (f" Â±{tol}%" if tol is not None else " (tol ?)")
        if tppm is not None: label += f" ({tppm}ppm/K)"
        cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)
        cv2.putText(img,label,(x1,max(20,y1-10)),cv2.FONT_HERSHEY_SIMPLEX,0.7,(0,0,255),2)

plt.figure(figsize=(10,10)); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.axis("off"); plt.show()
out_path = os.path.join(SAVE_DIR, f"out_{os.path.basename(img_path)}"); cv2.imwrite(out_path, img); print("Saved:", out_path)
```
