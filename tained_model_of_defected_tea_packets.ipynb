```python
import os, random, shutil

IMG_DIR = "tea_dataset/images"
LBL_DIR = "tea_dataset/labels"

train_img = "tea_dataset/images/train"
val_img   = "tea_dataset/images/val"
train_lbl = "tea_dataset/labels/train"
val_lbl   = "tea_dataset/labels/val"

for d in [train_img, val_img, train_lbl, val_lbl]:
    os.makedirs(d, exist_ok=True)

images = [f for f in os.listdir(IMG_DIR) if f.endswith(".jpg")]
random.shuffle(images)

split = int(0.8 * len(images))
train_imgs = images[:split]
val_imgs   = images[split:]

for f in train_imgs:
    shutil.move(f"{IMG_DIR}/{f}", f"{train_img}/{f}")
    shutil.move(f"{LBL_DIR}/{f.replace('.jpg','.txt')}", f"{train_lbl}/{f.replace('.jpg','.txt')}")

for f in val_imgs:
    shutil.move(f"{IMG_DIR}/{f}", f"{val_img}/{f}")
    shutil.move(f"{LBL_DIR}/{f.replace('.jpg','.txt')}", f"{val_lbl}/{f.replace('.jpg','.txt')}")

print("Dataset split completed")
```

```python
%%writefile tea_data.yaml
path: tea_dataset
train: images/train
val: images/val

nc: 2
names:
  0: defective
  1: non_defective
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # nano = best for Raspberry Pi

model.train(
    data="tea_data.yaml",
    epochs=100,
    imgsz=640,
    batch=8,
    device=0,
    workers=2,
    name="tea_packet_defect"
)
```

```python
model.val()
```

```python
model.predict(
    source="tea_dataset/images/val",
    conf=0.4,
    save=True
)
```

```python
model.export(format="onnx")
model.export(format="tflite")
```
